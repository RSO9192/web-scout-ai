```python
from web_scout.scraping import Crawl4AICrawler, Orchestrator, OrchestratorConfig

config = OrchestratorConfig(
    max_depth=1,
    max_urls=5,
    max_concurrent_fetches=2,
    max_concurrent_parses=2,
)
orch = Orchestrator(config, crawler=Crawl4AICrawler(max_links=3))

seed_url = "https://www.fao.org/faolex/results/details/en/c/LEX-FAOC243040"
await orch.queue_url(seed_url)
results = await orch.run()

print(f"Collected {len(results)} page(s) from seed {seed_url!r}")
for result in results:
    print(f"- {result.url} ({len(result.text_content)} chars, {len(result.links)} links)")
```
