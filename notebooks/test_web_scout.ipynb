# web-scout-ai — interactive test notebook

```python
import sys, asyncio, logging
from pathlib import Path
from dotenv import load_dotenv

# Load API keys
load_dotenv(Path('..') / '.env')

# Show pipeline logs
logging.basicConfig(level=logging.INFO, format='%(levelname)s %(name)s: %(message)s')
logging.getLogger('LiteLLM').setLevel(logging.WARNING)
logging.getLogger('litellm').setLevel(logging.WARNING)

from web_scout import run_web_research
from web_scout.agent import DEFAULT_WEB_RESEARCH_MODELS

print('Ready. Models:', DEFAULT_WEB_RESEARCH_MODELS)
```

## Configuration

```python
QUERY          = 'What are the main threats to coral reefs in Australia?'
MODELS         = DEFAULT_WEB_RESEARCH_MODELS
DEPTH          = 'standard' # 'standard' or 'deep'
```

## Run

```python
result = await run_web_research(
    query=QUERY,
    models=MODELS,
    research_depth=DEPTH,
    allowed_domains=["fao.org"],
)
```

## Results

```python
print('=' * 60)
print('SYNTHESIS')
print('=' * 60)
print(result.synthesis)
```

```python
print(f'SCRAPED SOURCES ({len(result.scraped)})')
for s in result.scraped:
    print(f'  - {s.title or s.url}')
    print(f'    {s.url}')

if result.scrape_failed:
    print(f'\nFailed: {len(result.scrape_failed)}')
    for s in result.scrape_failed:
        print(f'  - {s.url}')
```

```python
print(f'QUERIES ({len(result.queries)} executed)')
for q in result.queries:
    print(f'  - {q.query}')
```

## Inspect a source's extracted content

```python
# Change index to inspect a different source
idx = 0
if result.scraped:
    s = result.scraped[idx]
    print(f'Title : {s.title}')
    print(f'URL   : {s.url}')
    print('-' * 60)
    print(s.content)
```
